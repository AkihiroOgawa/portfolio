```python
import os
# 現在のディレクトリを取得
current_dir = os.getcwd()
print(f"現在のディレクトリ: {current_dir}")
print("内容一覧:")


for item in os.listdir(current_dir):
    full_path = os.path.join(current_dir, item)
    if os.path.isdir(full_path):
        print(f"[フォルダ] {item}")
    else:
        print(f"[ファイル] {item}")
```
